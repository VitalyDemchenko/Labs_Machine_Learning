```python
# Завдання 7 Визначте максимальну, мінімальну, медіанну ціну за
# допомогою методів  max, min, mean, median
max_price = data['price'].max()
min_price = data['price'].min()
mean_price = data['price'].mean()
median_price = data['price'].median()

print("Максимальна ціна = " + str(max_price))
print("Мінімальна ціна = " + str(min_price))
print("Середня ціна = " + str(mean_price))
print("Медіанна ціна = " + str(median_price))
```

```python
# Завдання 8 Виконайте наступні завдання з пошуку і сортування даних
print()
print("Відповіді на питання до завдання 8 " )
print()

# Питання: Який рейтинг у датасеті найвищий? 
rating_up = data['user_rating'].max()
print("Питання: Який рейтинг у датасеті найвищий? ")
print("Відповідь: " + str(rating_up))
print()

# Питання: Скільки книг мають такий рейтинг? 
books_whighest = data[data['user_rating'] == rating_up]
books_whighest_count = books_whighest.shape[0]
print("Питання: Скільки книг мають такий рейтинг? ")
print("Відповідь: " + str(books_whighest_count))
print()

# Питання: Яка книга має найбільше відгуків? 
index_max_review = data['reviews'].idxmax()
book_max_review = data.loc[index_max_review, 'name']
print("Питання: Яка книга має найбільше відгуків? ")
print("Відповідь: " + str(book_max_review))
print()

# Питання: З тих книг, що потрапили до Топ-50 у 2010 році, яка книга найдорожча ?
top50_2010_exp = data[(data['year'] == 2010)]
most_exp_2010 = top50_2010_exp[top50_2010_exp['price'] == top50_2010_exp['price'].max()]
print("Питання: З тих книг, що потрапили до Топ-50 у 2010 році, яка книга найдорожча ? ")
print("Відповідь: " + str(most_exp_2010))
print()

# Питання : Скільки книг жанру Fiction потрапили до Топ-50 у 2012 році 
fict_books = data[(data['genre'] == 'Fiction') & (data['year'] == 2012)]
count_fict = fict_books.shape[0]
print("Питання : Скільки книг жанру Fiction потрапили до Топ-50 у 2012 році ")
print("Відповідь: " + str(count_fict))
print()

# Питання : Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або метод isin)? 
book_rt49 = data[(data['user_rating'] == 4.9) & (data['year'].isin([2010,2011]))]
count_rt49 = book_rt49.shape[0]
print("Питання : Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або метод isin)?  ")
print("Відповідь: " + str(count_rt49))
print()
```

```python
# Завдання 9 Вивести максимальну і  мінімальну ціну для жанру Fiction
# I NonFiction (за допомогою методів  groupby та agg, для підрахунку мінімальних 
# та максимальних значень використовуйте max та min). Використовуйте тількі потрібні вам стовпці.
print()
res = data[data['genre'].isin(['Fiction', 'NonFiction'])].groupby('genre')['price'].agg(['min', 'max'])
print(res)
print()
```
