```python
import pandas as pd
# Завдання 1 Прочитайте CSV файл (використовуйте метод read_csv)

data = pd.read_csv("yt_statistic.csv", encoding = 'latin1')
data
```

```python
# Завдання 2 Виведіть перші п'ять рядків (використовується метод head)

print(data.head())
```

```python
# Завдання 3 Виведіть розміри датасету (використовуйте метод shape)

print("Розмір = ", data.shape )
```

```python
# Завдання 4  Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків
# у кожному зі стовпців (використовуйте методи isna та sum)

miss_data = data.isna().sum()/len(data)*100
print(miss_data)
```

```python
 # Завдання 5 Замініть комірки з пропущеними значенями на середні за стовпцем. 
    # І тип даних на float
    
data = data.fillna(data.mean())
data['subscribers'] = data['subscribers'].astype(float)
data['video views'] = data['video views'].astype(float)
data['uploads'] = data['uploads'].astype(float)
data
```

```python
# Завдання 6 Ще раз перевірте кількість пропусків.

miss_data2 = data.isna().sum()/len(data)*100
print(miss_data2)
```

```python
# Завдання 7 Перевірте, скільки унікальних країн представлено у колонці "Country"

un_countries = data['Country'].nunique()
print(" унікальних країн = " + str(un_countries))
```

```python
# Завдання 8 Побудуйте діаграму розподілу переглядів (використовуйте kind='hist')

import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.hist(data['video views'], bins = 25, color= 'red', edgecolor= 'black')
plt.title('Розподіл переглядів')
plt.xlabel('К-ть переглядів')
plt.ylabel('Частота')
plt.grid(axis = 'y', alpha = 0.75)
plt.show()
```

```python
# Завдання 9 Визначте максимальну, мінімальну, середню кількість переглядів
# (використовуйте методи max, min, mean)

max_view = data['video views'].max()
min_view = data['video views'].min()
mean_view = data['video views'].mean()

print("Максимальна к-ть =  " + str(max_view))
print("Мінімальна к-ть переглядів =  " + str(min_view))
print("Середня к-ть =  " + str(mean_view))
```

```python
# Завдання 10 Знайдіть країну, де найбільше відео було завантажено на YouTube

country_count = data.groupby('Country')['uploads'].sum()
country_most_viewed = country_count.idxmax()
most_vid = country_count.max()
print("Країна з найбільшою к-тю відео - " + str(country_most_viewed))
print("Кількість відео = " + str(most_vid))
```

```python
# Завдання 11 Знайдіть відео (виведіть його назву) з найбільшою
# та найменшою кількістю завантажень uploads.
max_upl = data['uploads'].idxmax()
vid_max = data.loc[max_upl, 'Title']
min_upl = data['uploads'].idxmin()
vid_min = data.loc[min_upl, 'Title']
print("Найбільша к-ть завантажень " + str(vid_max))
print("Найменша к-ть завантажень " + str(vid_min))
```
