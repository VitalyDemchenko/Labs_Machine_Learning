```python
# Завдання 1 Виведіть перші 5 рядків таблиці

import pandas as pd
import numpy as np

# Зчитуємо дані з вказаного URL
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

# Зчитуємо таблицю в DataFrame
data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця

df_1 = data.head(5)
# Виводимо перші 5 рядків таблиці
print(df_1)


```

```python
# Завдання 2 Визначте кількість рядків і стовпчиків за допомогою атрибуту shape

num_rows, num_columns = data.shape
print(f"Кількість рядків: {num_rows}, Кількість стовпчиків: {num_columns}")
```

```python
# Завдання 3 Визначте загальну кількість років та кількість місяців у даних.

num_years = len(data['Рік']) - 1  
num_months = num_years*12  
print("Загальна кількість років =", num_years)
print("Загальна кількість місяців =", num_months)
```

```python
# Завдання 4 Замінити пропущені значення  «-» в таблиці на NaN

data = data.rename(columns={
    "Рік": "Year",
    "В національній валюті": "Salary in UAH",
    "В доларах США": "Salary in USD",
    "Кількість кг хліба[20] на середню зарплату": "Bread per Salary",
    "Кількість БігМаків на середню зарплату": "BigMacs per Salary"
})
data = data.replace("—", np.nan)
data
```

```python
# Завдання 5 Визначити тип стовпців за допомогою dtypes

column_types = data.dtypes
print("Типи стовпців:")
print(column_types)
```

```python
# Завдання 6 Видаліть передостанній і останній стовпчики ( Кількість кгхліба[20] на середнюзарплату,
# Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею. 

data = data.drop(columns=["Bread per Salary", "BigMacs per Salary"])
data
```

```python
# Завдання 7 Замінити тип нечислових даних на числові
data.dtypes
data["Year"] = data["Year"].astype(str).str.replace("*", "")
data["Year"] = data["Year"].astype(str).astype(int)

data["Salary in UAH"] = data["Salary in UAH"].astype(str).str.replace('▲', "").str.replace('▼', "")
data["Salary in USD"] = data["Salary in USD"].astype(str).str.replace('▲', "").str.replace('▼', "")

data['Salary in UAH'] = data['Salary in UAH'].astype(str).str.replace(',', '.').astype(float)
data['Salary in USD'] = data['Salary in USD'].astype(str).str.replace(',', '.').astype(float)
data
data.dtypes
```

```python
data
```

```python
# Завдання 8 Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len)

miss_vallue = data.isnull()
miss_count = miss_vallue.sum()
total_rows = len(data)
percent_of_miss = (miss_count / total_rows) * 100
print(percent_of_miss)
```

```python
# Завдання 9 Замініть всі пропущені значення на середні значення по (стовпцю)
data = data.fillna(data.mean())
data
```

```python
# Завдання 10 Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату.

highest_sal_year = data[data['Salary in UAH'] == data ['Salary in UAH'].max()]['Year'].values[0]
highest_salary = data['Salary in UAH'].max()
print ("Рік з найвищою зарплатою " + "(" + "В гривнях" + ") = " +  str(highest_sal_year)  )
print ("Зарплата в цьому році = " + str(highest_salary) )
```

```python
# Завдання 11 Знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату

min_sal_year_dol = data[data['Salary in USD'] == data ['Salary in USD'].min()]['Year'].values[0]
min_salary_dol = data['Salary in USD'].min()
print ("Рік з найнищою зарплатою " + "(" + "В доларах" + ") = " +  str(min_sal_year_dol)  )
print ("Зарплата в цьому році = " + str(min_salary_dol) )
```

```python
# Завдання 12 Виведіть роки, коли зарплата в грн була вище середньої

average_salary = data['Salary in UAH'].mean()
print("середня зарплата")
print(average_salary)
above_average = data[data['Salary in UAH'] > average_salary]['Year']
print("роки, коли зарплата в грн була вище середньої")
print(above_average)
```

```python
# Завдання 13 Порівняйте  зарплату  в грн за останні два доступних роки.

last2 = data['Year'].sort_values(ascending = False).head(2)
salary = data[data['Year'].isin(last2)]['Salary in UAH']
print(salary)
difference = salary.iloc[0] - salary.iloc[1]

print("Різниця між зарплатами = " + str(difference))
```

```python
# Завдання 14 Створіть графік, що відображає  зарплату в Україні з часом (за роками).

import matplotlib.pyplot as plt

years = data['Year']
salaries = data['Salary in UAH']
plt.figure(figsize = (10,10))
plt.plot(years,salaries, marker = 'o', linestyle = '-')
plt.title("Зарплата в Україні з 1991")
plt.xlabel('Рік')
plt.ylabel('Розмір зарплати')
plt.grid(True)
plt.show()
```

```python
# Завдання 15 Видаліть останній рік з таблиці та виведіть оновлену таблицю.

new_table_no_last_year = data.drop(data.index[-1])
new_table_no_last_year
```

```python
# Завдання 16 Побудуйте графік, що відображає середню зарплату в Україні 
# в розрізі років без останнього року.

years = new_table_no_last_year['Year']
salaries = new_table_no_last_year['Salary in UAH']
plt.figure(figsize = (10,10))
plt.plot(years,salaries, marker = 'o', linestyle = '-')
plt.title("Зарплата в Україні з 1991")
plt.xlabel('Рік')
plt.ylabel('Розмір зарплати')
plt.grid(True)
plt.show()

```

```python
# Завдання 17 Обчисліть зарплату за останні два роки і порівняйте її 
# із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку.

last_two = data[data['Year'].isin(data['Year'].nlargest(2))]

last_five = data[data['Year'].isin(data['Year'].nlargest(5))]
salary_last_two = last_two['Salary in UAH'].sum()

salary_last_five = last_five['Salary in UAH'].sum()

difference = salary_last_two - salary_last_five

if difference > 0:
    text = "збільшилась"
elif difference < 0:
    text = "зменшилась"
else:
    text = "не змінилась"
print("Зарплата")
print("За останні 2 роки = " + str(salary_last_two))
print("За останні 5 років = " + str(salary_last_five))
print("Різниця = " + str(difference))
print("Отже, зарплата за ці роки " + str(text))
```
